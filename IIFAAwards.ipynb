```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

url="https://en.wikipedia.org/wiki/International_Indian_Film_Academy_Awards"
page=requests.get(url)
soup=BeautifulSoup(page.content,'html5lib')
```

```python
tdata=soup.find(class_="sortable wikitable")
#tdata
```

```python
heads=[g.get_text().strip() for g in tdata.find_all('th')]
#heads
```

```python
data=[[g.get_text().strip() for g in t.find_all('td')] for t in tdata.find_all('tr')[1:]]
# data
# for i in data:
#     print(len(i))
```

```python
#data1=[t.find_all('td')[3].get_text().replace('\n','') for t in tdata.find_all('tr')[1:]]
# #data1

# for j in data:    
#     if len(j)==8:
#         cont=j[-1]
#         print("len=8",cont)
#     elif len(j)==7:
#         host=j[3]
#         print("len=7",host)
#     elif len(j)==6:
#         print("len=6",host,cont)
        
d={k:[] for k in heads}

for p in data:     
    
    if len(p)==8:   
        cont=p[-1]
        d['Ceremony'].append(p[0])
        d['Date'].append(p[1])
        d['Best Film winner'].append(p[2])      
        d['Host(s)'].append(p[3])    
        d['Venue'].append(p[4])
        d['City'].append(p[5])
        d['Country'].append(p[6])
        d['Continent'].append(p[7])
        
    elif len(p)==7:
        host=p[3]
        d['Ceremony'].append(p[0])
        d['Date'].append(p[1])
        d['Best Film winner'].append(p[2])
        d['Host(s)'].append(p[3])            
        d['Venue'].append(p[4])
        d['City'].append(p[5])
        d['Country'].append(p[6])
        d['Continent'].append(cont)
    
    elif len(p)==6:
        d['Ceremony'].append(p[0])
        d['Date'].append(p[1])
        d['Best Film winner'].append(p[2])
        d['Host(s)'].append(host)            
        d['Venue'].append(p[3])
        d['City'].append(p[4])
        d['Country'].append(p[5])
        d['Continent'].append(cont)
    
    elif len(p)==4:
        d['Ceremony'].append(p[0])
        d['Date'].append(p[1])
        d['Best Film winner'].append(p[2])      
        d['Host(s)'].append(p[3])    
        d['Venue'].append(None)
        d['City'].append(None)
        d['Country'].append(None)
        d['Continent'].append(None)

df=pd.DataFrame(d)
df
```
